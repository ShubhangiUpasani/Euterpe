```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
import numpy as np
import pandas as pd
import os
import pickle
import csv
from matplotlib import pyplot as plt
import seaborn as sns
import pickle
import re
DATA_DIR = 'data/'
```

```python
with open(DATA_DIR+"metrolyrics_languages.pickle","rb") as f:
    data=pickle.load(f)
    
    
lyrics_data=data["lyrics"]
print(data.shape)
```

```python
print(lyrics_data)
processed_lyrics = []
for l in lyrics_data:
    processed_lyrics.append(np.char.replace(re.sub("[\(\[].*?[\)\]]", "", l), '\n', ' '))
    
processed_lyrics = np.array(processed_lyrics)
print(processed_lyrics[89])
```

```python
print(processed_lyrics.shape)
data.columns
print(data.shape)
titles = data['title']
titles = np.array(titles)
genre = data['genre']
genre = np.array(genre)
```

```python
# for i, doc in enumerate(processed_lyrics):
#     tag = (titles[i], data.loc[int(i), 'genre'])
#     print(tag)
documents=[TaggedDocument(words=word_tokenize(doc.lower()),tags=[(titles[i], genre[i])]) for i, doc in enumerate(processed_lyrics)]
```

```python
max_epochs = 100

alpha = 0.025

model = Doc2Vec(alpha=0.025,
                min_alpha=0.025,
                workers=15, 
                min_count=2,
                window=10,
                size=300,
                iter=20,
                sample=0.001,
                negative=5)
```

```python
model.build_vocab(documents)
epochs = 10
for epoch in range(epochs):
    print('epoch', epoch)
    model.train(documents, total_examples=model.corpus_count, epochs=model.iter)
    model.alpha -= 0.002  # decrease the learning rate
    model.min_alpha = model.alpha  # fix the learning rate, no decay
```
