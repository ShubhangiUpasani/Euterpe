```python
with open('data/lyrics_genre_MSD_1000chunks.pickle', 'rb') as chunkf:
    df = pickle.load(chunkf)
```

```python
df.head()
```

```python
# import unicode
allLyrics = df['lyrics'][0:20000]
combo=""
for lyric in allLyrics:
    lyric = str(lyric)
    lyric = re.sub(r'\W+ \n', '', lyric)
    combo = combo+lyric+"\n\n"
    combo = combo.encode().decode('utf-8','ignore')
#     combo = unicode(combo, "utf-8")
#     decode('utf-8', 'ignore')
print(combo)
with open('allLyrics.txt','a') as f:
    f.write(combo)
    f.close()
```
