```python
import numpy as np
import pandas as pd
import os
import pickle
import csv
from matplotlib import pyplot as plt
import seaborn as sns
import pickle
DATA_DIR = 'data/'
```

```python
with open(DATA_DIR+"lyrics_genre_MSD_1000chunks.pickle","rb") as f:
    data=pickle.load(f)
    
    
data=data["lyrics"]
print(data.shape)
```

```python
from gensim.models.doc2vec import Doc2Vec,TaggedDocument
documents=[TaggedDocument(doc,[i]) for i, doc in enumerate(data)]
```

```python
model=Doc2Vec(documents,alpha=0.025,min_alpha=0.025,workers=15,min_count=2,window=10,size=300,iter=20,sample=0.001,negative=5)

#explanation for each argument - https://tmthyjames.github.io/2018/january/Analyzing-Rap-Lyrics-Using-Word-Vectors/
```

```python
model.docvecs[97].shape # one feature vector
```

```python
model.docvecs[97]
```

```python
feature_vectors=[]

for i in range(data.shape[0]):
    feature_vectors.append(model.docvecs[i])
    
feature_vectors=np.asarray(feature_vectors)
feature_vectors.shape
```

```python
np.save(DATA_DIR+"feature_vectors.npy",feature_vectors)
```
